# Titanic Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\ocdga\Downloads\archive (1)\Titanic-Dataset.csv")
```

# 5 rows

```python
print(df.head())
```

# Adults (Age >= 18) and Fare > 30

```python
filtered_df = df[(df['Age'] >= 18) & (df['Fare'] > 30)]
print(filtered_df)
```

```python
print("\nFiltered Passengers: ")
print(filtered_df[['Name','Age','Fare']].head())
```

Family Size
# SibSp = siblings/spouses
# Parch = parents/children

```python
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
print("\nDataFrame with Family Size:")
print(df[['Name', 'SibSp', 'Parch', 'Family_Size']].head())
```

```python
df['Fare_Per_Person'] = df['Fare'] / df['Family_Size']


print("\nNew Features Added:")


print(df[['Family_Size', 'Fare_Per_Person']].head())
```

# Groupby analysis

```python
group_analysis = df.groupby(['Pclass', 'Sex']).agg({
    'Fare': 'mean',
    'Survived': 'mean'
}).rename(columns={
    'Fare': 'Average_Fare',
    'Survived': 'Survival_Rate'
})

print("\nGroupby Analysis:")
print(group_analysis)
```
